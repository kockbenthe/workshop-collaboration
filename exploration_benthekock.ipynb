#### Correlation heatmap for top features
What this does:
- find the 10 features most correlated (positively or negatively) with the target
- builds a smaller correlation matrix for just those columns
- draws a colorful heatmap showing strength and direction
    - red = positive correlation
    - blue = negative correlation
    - numbers = exact correlation values

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Compute correlations (numeric columns only)
corr = df.corr(numeric_only=True)

# Get top 10 features most correlated with target (by absolute value)
top_features = (
    corr['target']
    .abs()
    .sort_values(ascending=False)
    .head(11)  # includes 'target' + top 10 features
    .index
)

# Filter correlation matrix for just those features
corr_top = corr.loc[top_features, top_features]

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_top,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    cbar=True,
    square=True
)
plt.title("Correlation Heatmap – Target and Top Features", fontsize=14)
plt.tight_layout()
plt.show()

```

#### Check if top features are highly correlated with each other
If so, this means they contain redundant information. This leads to:
- **multicollinearity**: features explain the same variance, making model coefficients unstable.
- **overfitting risk**: the model "overcounts" the same signal.
- **wasted dimensionality**: more features = more complexity, slower training & less interpretability.

*What should we do?*
- keep one of them
- drop the others - they don't add unique information

```python
# -- Check for Collinearity

top_features_excl = top_features.drop('target')
# Suppose 'top_features' is your list of top correlated features
corr_matrix = df[top_features_excl].corr().abs()

# Show correlations between features (excluding self-correlation)
upper = corr_matrix.where(~np.tril(np.ones(corr_matrix.shape)).astype(bool))

# Find features highly correlated with others (e.g., > 0.85)
high_corr_pairs = [
    (col, row)
    for col in upper.columns
    for row in upper.index
    if pd.notnull(upper.loc[row, col]) and upper.loc[row, col] > 0.7
]
print("Highly correlated feature pairs:", high_corr_pairs)
```

Conclusion: no highly correlation among featurepairs. Continue.

### **Question 3: Turn the source data into new data file**

Turn the source data into a new data file that contains only the target variable and a few of the feature variables that are useful (at least 5). Save this data file in data/processed.

#### Calculate Correlations

```python
# -- Compute correlations with the target
print("\n--- Top 10 most positively correlated with target ---")
print(correlations.head(10))   # Top 10 most positively correlated

print("\n--- Top 10 most negatively correlated with target ---")
print(correlations.tail(10))   # 10 most negatively correlated
```

#### Pick the top 5-10  and Create reduced dataset
Usually, take the top N features with the largest absolute correlation values — not just positive ones. And create reduced dataset existing of these 5 features and the target value.

```python
# -- Select top 5 features
# Get top 5 features by absolute correlation
top_features = correlations.abs().sort_values(ascending=False).head(6)
print("\n--- Top 5 features by absolute correlated with target ---")
#print(top_features)

# This includes "target" as well. Drop "target":
useful_features = top_features.drop('target').index.tolist() # list
print(useful_features)



# -- Create new Dataset
# Keep only target and top 5 of selected features
df_reduced =pd.DataFrame(df[["target"] + useful_features])
df_reduced
```
